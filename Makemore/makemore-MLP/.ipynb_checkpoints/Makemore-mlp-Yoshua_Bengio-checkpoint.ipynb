```python
import torch
import matplotlib.pyplot as plt
```

```python
names  =open("names.txt","r").read().splitlines()
names[:8]
```

```python
chars = sorted(list(set("".join(names))))
string_int = {s:i+1 for i,s in enumerate(chars)}
string_int["."]=0
int_string = {s:i for i,s in string_int.items()}
int_string
```

```python
#now making the trigram
block_size = 3
x,y = [],[]
for i,name in enumerate(names):
    context = [0]*block_size
    for ch in name+".":
        ix =string_int[ch]
        x.append(context)
        y.append(ix)
        if i<5:
            print("".join(int_string[i] for i in context),"--------->",int_string[ix])
        context = context[1:]+[ix]

x = torch.tensor(x)
y= torch.tensor(y)
x,y
```

```python
x.dtype,x.shape,y.dtype,y.shape
```

```python
#archictecture of neural network
```

```python
# in this as we can see we take 3 previous words/character and get the 4th word
#how it works (the base idea): - so there is a matix of C =  V x D V is vocabulary and D is Dimensions 
# vocabulary eg:- 
# ["the","dog","is","in","the","room"] the vocabualry is ["the","dog","is","in","room"] 
# we have to embed the V(vocabularies ) into a D column aka dimension so eg "the dog is in the room" the C will be 4x2 (4 cause of 4 words in the vocabulary and 2 which is dimsnsion is fully onto the users input as much as he want )
# then we'll use backpropagation to make those closer which are mostly common by MLP
# but in case of name prediction our vocabular will be 27 casue there are 27 character
```

```python
#A more thorough   explanation for the input layer(C) i can give you is that 
# as we know that the x which is input is consist of 1x3 [n,n,n] array and each n in that array is 
# a character which has a output y so to embedd it 
# we'll have to first make a 27xk(any number you want) matrix then
# for x we'll have to make it one-hot encoding so each array which is consit of 3 character will be turned into 
# [0,0..->(27),1] and will be turned into 3x27 marix of one hot encoding and if there is a word then (no of character in that word)x3x27
# and then each row from 3x27 one hot encoding goes to the three 3 input layes as you can see on the image 
```

```python
# and there is a bit clarification the matrix we've all studied in school is bit differ from this cause in school the rows consist of {x,y,z} and columns consist of {x1,x2,x3..} where as in this  the rows are {x1,x2,x3..->(dependent on the no of vocabulary)} and column are {x,y,z..->(dimension as you wish)}
```

```python
#making the embedding table C
# As we know there are 27 possible character a,b,c->z,. and we can make column/dimensions as much as we want 
C = torch.randn((27,2))
C
```

```python
#and as we've discussed that we'll have to turn each 1x3 matrix of 228146x3 of x into one-hot encoding 
#so here's how to do that 
import torch.nn.functional as F
F.one_hot(torch.tensor(5),num_classes=27).float() # 5 is just for example and a int because string to integre/str_int
```

```python
F.one_hot(torch.tensor(5),num_classes=27).float() @ C
```

```python
# if we can see the output of this dot product we can notice it is the same as the 5th row of the C matrix so we can say that 
# one-hot of each word @ C == C[the strint to integre/str_int]
C[5],C[5]== F.one_hot(torch.tensor(5),num_classes=27).float() @ C# as we can see both are same
```

```python
emb = C[x]#the same dot product or can say emedding of one vector of x in C
emb[:10] ,emb.shape
```

```python
#Now we have the 1st layer its time to make the 2nd layer 
# so before making the 2nd layer we'll have to know how many input does the 2nd layer takes
# as we can see from the C[X]/emb matrix there is 3 vector with 2 dimension each so the input to the 2nd layer will be 2*3 =6
W1 = torch.rand((6,100))#created weight 
b1 = torch.rand(100)# created bias
```

```python
#before doing dot-product with W1 we'll have to turn the emb matrix which is of size 32x3x2 in to 32x6 
# there are manny ways of doing it i can tell you a few and you can choose any of that
```

```python
#1st way
torch.cat([emb[:,0,:],emb[:,1,:],emb[:,2,:]],1).shape
#but we're not gonna use it because what if we change the 1st input layer instead of  3 we take  4 or 3 we take 2 we'll have to come again to this cell and change the code 
```

```python
#2nd way
torch.cat(torch.unbind(emb,1),1).shape
#we can use this but there is 1 more which is more efficent and effective way
```

```python
#3rd way
emb = emb.view(emb.shape[0],6)
emb[:10,:],emb.shape
# view function transforms any type of vector into desired vector as long as 
# so we are going to use the 3rd method 
```

```python
#Now multiplying it and using actiavtion function
h = torch.tanh(emb@W1+b1)
h[:1,:],h.shape
```

```python
# now going to the 3rd and final layer 
W2 = torch.randn([100,27])
b2 = torch.randn(27)
```

```python
logits = h@W2+b2
logits.shape
```

```python
#Now using the cross entrophy activation function
counts = logits.exp()
prob = counts/counts.sum(1,keepdims=True)
prob[0].sum()
```

```python
# Now as we've seen in the biagram makemore we take the probablity ditribution of a array of this cross entrophy solution and then take the Y and use gradient decent to increase the probablity at that point in the matrix of cross entrophy
#what i mean that we have to increase this 
prob[torch.arange(y.shape[0]),y][:3]# i put this [:3] because i dont want to show all the 228146 elements in one so thats why anyways you can use the .shape to see it 
```

```python
# now making the loss function 
loss = -prob[torch.arange(y.shape[0]),y].log().mean()
loss
```

```python
#or insted of using maniual cross entrophy we can use 
F.cross_entropy(logits,y)# and you can see the loss fucntion result will be the same
```

```python
#_________________Now organaizing everyting in a neat manner the forward pass and the backward pass______________#
```

```python
x.shape,y.shape
```

```python
C= torch.randn(27,2)
W1 = torch.randn((6,100))
b1 = torch.randn(100)
W2 = torch.randn(100,27)
b2 = torch.randn(27)
parameters = [C,W1,b1,W2,b2]
```

```python
sum(p.nelement() for p in parameters) 
```

```python
for p in parameters:
    p.requires_grad=True
for _ in range(100):
    #forward pass
    emb = C[x]
    h = torch.tanh(emb.view(x.shape[0],6)@W1+b1)
    logits = h@W2 +b2

    #backward pass
    loss = F.cross_entropy(logits,y)

    for p in parameters:
        p.grad =None
    loss.backward()
    for p in parameters:
        p.data += -0.1*p.grad
print(f"loss : {loss.item()}")
```

```python
# Look from the begining we've been using the simple split the train split the test split 
#but now im going to introduce a new one and the reason on that is below :-
# when we train the data to much and the loss become somewhere 0 you will think its good but actually what its doing is its memorizing the data thats it nothing more but if we give it some new point the loss will be exponesial because model dose'nt know anything about data all what it was doing was memorizing the point its like over fitting but even regularization can't help us cause 
#regularization is pprevent it from overfitting by some points but we want the model to actually learn when new data is given we want the model to predict some common gound b/w its predicted value and the real predicted value 
# in that case We use another split called validation split to train the model in a better way {traning_split:80% test_split:10% validation_split: 10%} how it works is below :-
# so by far we know that the 80% of the data is taken by traning_split it uses all the parameters which and use there grad and data to decrease the loss
#but in validation_split we take that trained model and use something called hyper parameters 
# hyper-parameter are nothing but the thing which we can create to decrease the loss the more through explanation is :-
#Look at the layer 2nd we use W1(6,100) matrix  we could have use 99 columns insted of 100 and still there would've be no problem with that and thats what hyper parametrs is in case of increasing W1's column we will be increasing the number of nuerons thats it and the model will just improve and the loss will go down 
#thats what we do in validation split we use it like gradient discent but insted of changing the p.data we change the p or can say increase and decrese the parameter to help model to understand the data instead of memorizing all of it 
```

```python
def build_database(names):
    #now making the trigram
    block_size = 3
    x,y = [],[]
    for i,name in enumerate(names):
        context = [0]*block_size
        for ch in name+".":
            ix =string_int[ch]
            x.append(context)
            y.append(ix)
            context = context[1:]+[ix]

    x = torch.tensor(x)
    y= torch.tensor(y)
    print(x.shape,y.shape)
    return x,y
import random 
random.seed(42)
random.shuffle(names)
n1 = int(0.8*len(names))
n2 = int(0.9*len(names))
xtr,ytr = build_database(names[:n1])
xdev,ydev = build_database(names[n1:n2])
xte,yte = build_database(names[n2:])
```

```python
#____________________________by using the validation split_______________________________#
```

```python
C= torch.randn(27,2)
W1 = torch.randn((6,100))
b1 = torch.randn(100)
W2 = torch.randn(100,27)
b2 = torch.randn(27)
parameters = [C,W1,b1,W2,b2]
```

```python
for p in parameters:
    p.requires_grad=True
for _ in range(100):
    #forward pass
    emb = C[xtr]
    h = torch.tanh(emb.view(xtr.shape[0],6)@W1+b1)
    logits = h@W2 +b2

    #backward pass
    loss = F.cross_entropy(logits,ytr)
    for p in parameters:
        p.grad =None
    loss.backward()
    for p in parameters:
        p.data += -0.1*p.grad
print(f"loss : {loss.item()}")
```

```python
emb = C[xdev]
h = torch.tanh(emb.view(xdev.shape[0],6)@W1+b1)
logits = h@W2 +b2
loss = F.cross_entropy(logits,ydev)
loss
```

```python
#Now sampling from the model
g = torch.Generator().manual_seed(1123456789)
for _ in range(20):
    out = []
    context = [0]*block_size
    while True:
        emb =C[torch.tensor([context])]
        h = torch.tanh(emb.view(1,-1) @ W1+b1)
        logits = h@W2+b2
        probs = F.softmax(logits,dim=1)
        ix = torch.multinomial(probs,num_samples=1,generator=g).item()
        context =context[1:]+[ix]
        out.append(ix)
        if ix ==0:
            break
    print("".join(int_string[i]for i in out))
        
```
